```python
import pandas as pd
import re
```

```python
df = pd.read_csv("~/daft.ie_dataengineering_and_analysis/daft.ie_dataengineering_and_analysis/include/data_v1.csv")
```

```python
df['COUNTY'] = df.Location.apply(lambda x : re.split('[._]', x)[-1])
```
